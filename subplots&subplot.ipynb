````python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
%matplotlib inline
````

````python
x = np.linspace(-np.pi,np.pi,100)
y1 = np.tanh(x)
y2 = np.exp(x)
y3 = np.sin(np.pi*x)
y4 = 1/x
````

````python
plt.figure(figsize=(20,8),dpi=100)
plt.plot(x,y1,label='tan(x)')
plt.plot(x,y2,label='e^x')
plt.plot(x,y3,label='sin(x)')
plt.plot(x,y4,label='1/x')
plt.ylim(-2,5)
plt.title('this is x-y pics',fontsize=15)
plt.legend(fontsize=15)
plt.show()
````

## subplot

````python
plt.figure(figsize=(10,12),dpi=100)

plt.subplot(4,1,1)
plt.plot(x,y1,label='tan(x)')
plt.ylabel('y1 value')

plt.subplot(4,1,2)
plt.plot(x,y2,label='e^x')
plt.ylabel('y2 value')

plt.subplot(4,1,3)
plt.plot(x,y3,label='sin(x)')
plt.ylabel('y3 value')

plt.subplot(4,1,4)
plt.plot(x,y4,label='1/x')
plt.ylabel('y4 value')
````

````python
plt.figure(figsize=(12,12),dpi=100)

plt.subplot(221)
plt.plot(x,y1,label='tan(x)')
plt.ylabel('y1 value')

plt.subplot(222)
plt.plot(x,y2,label='e^x')
plt.ylabel('y2 value')

plt.subplot(223)
plt.plot(x,y3,label='sin(x)')
plt.ylabel('y3 value')

plt.subplot(224)
plt.plot(x,y4,label='1/x')
plt.ylabel('y4 value')
````

## add_subplot()

````python
fig = plt.figure(figsize=(12,12),dpi=100)

ax1 = fig.add_subplot(221)
ax1.plot(x,y1)

ax2 = fig.add_subplot(222)
ax2.plot(x,y2)

ax3 = fig.add_subplot(223)
ax3.plot(x,y3)

ax4 = fig.add_subplot(224)
ax4.plot(x,y4)
````
