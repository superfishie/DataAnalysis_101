```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])
```

```python
dates1 = pd.date_range('20170101',periods=20)
dates2 = pd.date_range('20170102',periods=21)
print dates1
print dates2
```

```python
df1 = pd.DataFrame(np.random.randn(20,4), index=dates1, columns=list('ABCD'))
df1.iloc[3,3]=np.nan
df1
```

```python
df2 = pd.DataFrame(np.random.randn(21,1), index=dates2, columns=['y'])
df2.iloc[7] = np.nan
df2
```

```python
frames=[df1,df2]
pd.concat(frames) #concat默认是上下堆积
```

### Concate

```python
pd.concat(frames,axis=1) #加上axis=1表示横向连接
```

```python
pd.concat(frames,axis=1,join='inner') #内联
```

```python
pd.concat(frames,axis=1,join_axes=[df2.index])
```

```python
def alignment_filter(df_x,y):
    df = pd.concat([df_x,y],axis=1,join='inner').dropna(how='any',axis=0)
    return df.loc[:,df_x.columns.values], df.loc[:,y.columns.values]
```

```python
x,y = alignment_filter(df1,df2)
```
