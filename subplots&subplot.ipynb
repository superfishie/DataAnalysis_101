```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-np.pi,np.pi,100)
y1 = np.tanh(x)
y2 = np.exp(x)
y3 = np.sin(np.pi*x)
y4
```

```python
plt.figure(figsize=(20,8),dpi=100)
plt.plot(x,y1,label='tan(x)')
plt.plot(x,y2,label='e^x')
plt.plot(x,y3,label='sin(x)')
plt.ylim(-2,5)
plt.title('this is x-y pics',fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

## subplot

```python
plt.figure(figsize=(10,12),dpi=100)

plt.subplot(3,1,1)
plt.plot(x,y1,label='tan(x)')
plt.ylabel('y1 value')

plt.subplot(3,1,2)
plt.plot(x,y2,label='e^x')
plt.ylabel('y2 value')

plt.subplot(3,1,3)
plt.plot(x,y3,label='sin(x)')
plt.ylabel('y3 value')
```
