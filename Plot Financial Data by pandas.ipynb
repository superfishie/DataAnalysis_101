```python
import pandas as pd
from mpl_finance import candlestick_ochl
from matplotlib.pylab import date2num
import matplotlib.pyplot as plt 
```

```python
stock_data = pd.read_csv('stock_day.csv')
```

```python
stock_data.head()
```

```python
help(candlestick_ochl) 
```

```python
stock_data.index
```

```python
#把index转换为时间类型
pd.to_datetime(stock_data.index)
```

```python
# 股票数据 按时间排序
stock_data = stock_data.sort_index()
# 获取float类型的时间

stock_data.index = pd.to_datetime(stock_data.index)

day_k = stock_data[['open','close','high','low']]

day_k['time'] = date2num(stock_data.index)
day_k = day_k[['time','open','close','high','low']]

#获取float时间,转换为
day_k.head()
```

### 日线

```python
# help(plt.subplots)
fig,axes = plt.subplots(nrows=1,ncols=1,figsize = (20,8),dpi=100)  #获取坐标系
candlestick_ochl(axes,day_k.values[0:100],colorup='r',colordown='g')
#x轴以时间类型显示
axes.xaxis_date()
plt.show()
```

### 周线

```python
stock_data.head(10)
```

```python
#按周重采样 相当于分组 
weekday = stock_data.resample('W').last() #一周内最后一天的数据,close代表一周内最后一天的收盘价
#周开盘价
weekday['open'] =  stock_data['open'].resample('W').first()
#周内最高价
weekday['high'] =  stock_data['high'].resample('W').max()
#周内最低价
weekday['low'] =  stock_data['low'].resample('W').min()

weekday.head()
```

```python
weekday = weekday.sort_index()
weekday.index = pd.to_datetime(weekday.index)
week_k = weekday[['open','close','high','low']]
week_k['time'] = date2num(weekday.index)
week_k = week_k[['time','open','close','high','low']]

#获取float时间,转换为
week_k.head()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize = (20,8),dpi=100)  #获取坐标系
candlestick_ochl(axes,week_k.values[0:100],colorup='r',colordown='g')
#x轴以时间类型显示
axes.xaxis_date()
plt.show()
```

## 移动平均法

### 简单移动平均线(SMA)



```python
#绘制日k线
fig,axes = plt.subplots(nrows=1,ncols=1,figsize = (20,8),dpi=100) 
candlestick_ochl(axes,day_k.values[0:200],colorup='r',colordown='g')
axes.xaxis_date()

#绘制5,20,60日移动平均线
pd.rolling_mean(stock_data['close'][:200],window=5).plot()

# pd.rolling_mean(stock_data['close'][:200],window=20).plot()

# pd.rolling_mean(stock_data['close'][:200],window=60).plot()

stock_data['close'][:200].rolling(window=10,center=False).mean().plot()

plt.show()
```

### 指数平滑移动平均线(EWMA)


\begin{split}y_0 &= x_0 \\
y_t &= (1 - \alpha) y_{t-1} + \alpha x_t,\end{split}

$\alpha =\frac{2}{s + 1}$

```python
stock_data['close'].head()
```

```python
#5日移动平均线
stock_data['close'].ewm(span=5,min_periods=0,adjust=False,ignore_na=False).mean().head()
```

```python
#验证算法
s=5
alpha = 1/3
y0 = 12.52
x1 = 12.7
x2 = 12.9
y1 = (1-alpha)*y0 + alpha*x1
y2 = (1-alpha)*y1 + alpha*x2
y1,y2
```

```python
#绘制日k线
fig,axes = plt.subplots(nrows=1,ncols=1,figsize = (20,8),dpi=100) 
candlestick_ochl(axes,day_k.values[0:200],colorup='r',colordown='g')
axes.xaxis_date()

#绘制5日EWMA,SMA
# pd.ewma(stock_data['close'][:200],span=5).plot(label ="5ewma")
# pd.rolling_mean(stock_data['close'][:200],window=5).plot(label = "5sma")

stock_data['close'][:200].ewm(span=5,min_periods=0,adjust=True,ignore_na=False).mean().plot()
stock_data['close'][:200].rolling(window=5,center=False).mean().plot()

plt.show()
```

### 移动方差和标准差

```python
fig,axes = plt.subplots(nrows=1,ncols=1,figsize = (20,8),dpi=100) 
# pd.rolling_std(stock_data['close'][:200],window=5).plot(label='std') #标准
# pd.rolling_var(stock_data['close'][:200],window=5).plot(label='var') #方差

stock_data['close'][:200].rolling(window=5,center=False).std().plot()
stock_data['close'][:200].rolling(window=5,center=False).var().plot()
```
